### KD Tree to Speed up

```python
def bubble_sort(distance,name,name2,name3):
    lists=distance
    length = len(lists)
    # First level traversal
    for index in range(length):
        #Second level traversal
        for j in range(1, length - index):
            if lists[j - 1] > lists[j]:
                # Swap the two data, where temp is not used because of the python feature tuple.
                lists[j - 1], lists[j] = lists[j], lists[j - 1]
                name[j-1],name[j] = name[j],name[j - 1]
                name2[j-1],name2[j] = name2[j],name2[j - 1]
                name3[j-1],name3[j] = name3[j],name3[j - 1]
    return name,name2,name3
```

```python
def Quick_judgment(centerlng,centerlat,lng,lat):
    if abs(centerlng-lng)<0.002:
        if abs(centerlat-lat)<0.002:
            return True
        else:
            return False
    else:
        return False
```

```python
# BASETYPE	SUBTYPE	CATEGORY
def get_grid_list(centerlng,centerlat):
    BASETYPE=[]
    SUBTYPE=[]
    CATEGORY=[]
    distance=[]
    for index2, row2 in new_df.iterrows():
        if Quick_judgment(centerlng,centerlat,row2['wgslng'],row2['wgslat']):
            length=geodesic((centerlat,centerlng),(row2['wgslat'],row2['wgslng'])).m
            if length<100:

                distance.append(length)
                BASETYPE.append(row2['BASETYPE'])
                SUBTYPE.append(row2['SUBTYPE'])
                CATEGORY.append(row2['CATEGORY'])

    num=len(BASETYPE)
    
    RBASETYPE=' '.join(bubble_sort(distance,BASETYPE,SUBTYPE,CATEGORY)[0])
    RSUBTYPE=' '.join(bubble_sort(distance,BASETYPE,SUBTYPE,CATEGORY)[1])
    RCATEGORY=' '.join(bubble_sort(distance,BASETYPE,SUBTYPE,CATEGORY)[2])
    
    return num,RBASETYPE,RSUBTYPE,RCATEGORY
```
